```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.random.exponential(scale=2,size=1000)
sns.histplot(data,kde=True)
plt.title("positively skewed")
plt.show()
```

```python
log_data=np.log1p(data)
sns.histplot(log_data,kde=True)
plt.title("after transform")
plt.show()
```

```python
#for sales dataset do time series visualisation
import pandas as pd
import numpy as np
data={'Date':['2025-12-06','2025-11-09','2025-07-09','2025-10-27','2025-08-13','2025-06-09','2025-05-08'],'sales':[150,130,156,170,180,160,100]}
df=pd.DataFrame(data)
df['Date']=pd.to_datetime(df['Date'])
plt.plot(df['sales'],df['Date'],marker="o",color="red")
plt.title("time series")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

```python
data={'Date':['2025-12-06','2025-11-09','2025-07-09','2025-10-27','2025-08-13','2025-06-09','2025-05-08'],'sales':[150,200,180,220,170,300,250]}
df=pd.DataFrame(data)
df['Date']=pd.to_datetime(df['Date'])
plt.figure(figsize=(10,6))
#named color
plt.plot(df['Date'],df['sales'],color="purple",label="labelled with purle color",linewidth=2)
#Hex color
plt.scatter(df['Date'],df['sales'],color="#FF5733",label="hex color",s=100)
#RGB tuple
plt.plot(df['Date'],df['sales'],color=(0.1,0.2,0.3),linestyle="--",label="RGB Tuple")
#Grayscale
plt.fill_between(df['Date'],df['sales'],color="0.85",alpha=0.3,label="Grayscale")
plt.title("usage of different color codes")
plt.xlabel("sales")
plt.ylabel("Date")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

```

```python
#partial transparency and jittering
marks = [70, 75, 80, 75, 90, 85, 70, 88, 85, 85,
         80, 82, 70, 75, 90, 85, 82, 80, 75, 85]
x=np.random.normal(loc=1,scale=0.5,size=len(marks))
plt.figure(figsize=(6,5))
plt.scatter(x,marks,alpha=0.5,color='dodgerblue',edgecolor='black',s=100)
plt.title("  ")
plt.tight_layout()
plt.grid(True)
plt.show()
```
